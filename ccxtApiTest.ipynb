```python
import ccxt as ccxt
import pandas as pd
from datetime import datetime
```

```python
# Extract, Transform, Load the hourly ETH price data
cex_x = ccxt.binanceus().fetch_ohlcv('ETH/USDT', '1h')

allcex_uts = [datetime.utcfromtimestamp((xi[0]/1000)).strftime('%Y-%m-%d %H:%M:%S') for xi in cex_x]
times = pd.DataFrame(allcex_uts)

allcex_vals = [xi[4] for xi in cex_x]
vals = pd.DataFrame(allcex_vals)

newdf = pd.concat([times, vals], axis=1)
newdf.columns = ['timestamp', 'ETH Close Price']
df = newdf

print(df)
```

```python
df = df.rename(columns={'timestamp': 'date'})
df = df.rename(columns=({'ETH Close Price': 'close'}))
df
```
